```python
def agente_pandas(pregunta, df, corpus):

    pregunta = pregunta.lower()

    if "columnas" in pregunta:
        return list(df.columns)

    elif "filas" in pregunta or "registros" in pregunta:
        return f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas."

    elif "estadisticas" in pregunta or "estadísticas" in pregunta:
        return df.describe()

    elif "buscar" in pregunta:
        palabra = pregunta.replace("buscar", "").strip()

        resultados = corpus[
            corpus.str.contains(palabra, case=False, na=False)
        ]

        return resultados.head(10)

    else:
        return "No entendí la pregunta. Podés preguntar: columnas, filas, estadísticas o buscar un valor."
```

```python
df = pd.read_csv("dataset_normalizado.csv")

df.head()
```

```python
from langchain_core.documents import Document

from langchain_community.vectorstores import FAISS

from langchain_community.embeddings import FakeEmbeddings

from langchain_mistralai import ChatMistralAI
```

```python
documents = []

for index, row in df.iterrows():

    texto = " | ".join([
        f"{columna}: {row[columna]}"
        for columna in df.columns
    ])

    doc = Document(
        page_content=texto,
        metadata={"fila": index}
    )

    documents.append(doc)

print(f"Se crearon {len(documents)} documentos.")
```

```python
embeddings = FakeEmbeddings(size=768)

print("Embeddings creados correctamente.")
```

```python
vectorstore = FAISS.from_documents(
    documents,
    embeddings
)

print("Base vectorial FAISS creada correctamente.")
```

```python
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 5}
)

print("Retriever creado correctamente.")
```

```python
from transformers import pipeline
```

```python
llm = pipeline(
    "text-generation",
    model="google/flan-t5-base"
)

print("Modelo local cargado correctamente.")
```

```python
llm = pipeline(
    "text-generation",
    model="distilgpt2"
)

print("Modelo local cargado correctamente.")
```

```python
def rag_respuesta(pregunta):

    docs = retriever.invoke(pregunta)

    contexto = "\n".join([
        doc.page_content
        for doc in docs
    ])

    prompt = f"""
    Contexto:
    {contexto}

    Pregunta:
    {pregunta}

    Respuesta:
    """

    respuesta = llm(
        prompt,
        max_new_tokens=100
    )

    return respuesta[0]["generated_text"]
```
